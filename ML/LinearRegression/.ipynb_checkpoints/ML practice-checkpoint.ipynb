```python
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("default")
```

# We can import out data frame but now we will create one. And to do that lets first create an `empty Data Frame`

```python
df = pd.DataFrame()
```

# Now lets `assign values` to the Data Frame

```python
df["Outer Dia(mm)"] = [200, 250, 315, 355, 450, 525]
df["Inner Dia(mm)"] = [174.9, 218.8, 275.6, 310.5, 393.8, 459.4]
```

```python
df
```

# Now lets plot it to see how it's gonna be

```python
plt.plot(df["Outer Dia(mm)"], df["Inner Dia(mm)"],"*-")
plt.title("Outer Dia(mm) vs Inner Dia(mm) graph")
plt.xlabel("Outer Dia(mm)")
plt.ylabel("Inner Dia(mm)")
plt.show()
```

# Its a linear relationship. so we are going to do a LinearRegression. And for that we are going to need `sklearn` model.

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

# The thing is `sklearn` expects `numpy` values not the `pandas` values. So we gotta `convert` em into `numpy` values. First lets check its present situation then we will convert if necessary

```python
type(df["Outer Dia(mm)"]), type(df["Inner Dia(mm)"])
```

# See~!! these are pandas values. So we need to convert into `numpy`. And lets assign variable (X,y)

```python
X = df["Outer Dia(mm)"].values
y = df["Inner Dia(mm)"].values
```

```python
type(X), type(y)
```

# Now lets `fit` the dataframe into the model

```python
model.fit(X,y)
```

# It gives an error and says 

## ValueError: Expected 2D array, got 1D array instead:
## array=[200 250 315 355 450 525].
## Reshape your data either using array.reshape(-1, 1) if your data has a single feature or array.reshape(1, -1) if it contains a single sample.

# So lets `convert our 1D X values into 2D`. First we'll check their claim

```python
X.shape, y.shape
```

```python
X.ndim, y.ndim
```

# So X, y both are 1D. ok lets `convert X ino 2D`

```python
X = X.reshape(-1, 1)
```

# Now check it

```python
X.shape, X.ndim
```

# Great!!! Now lets `fit` it

```python
model.fit(X, y)
```

# Now lets see. How well did it fit!!!

```python
y_pred = model.predict(X)
y_pred
```

```python
y
```

# Lets check the `Accuracy` 

```python
accuracy = model.score(X, y)
accuracy*100
```

# Now lets predict for our `desired data`like for 500mm

```python
model.predict([[500]])
```

# It follows y=mx+c equation. Lets derive em

```python
m = model.coef_
c = model.intercept_
m, c 
```

```python
m*500 + c
```

# Now lets find the R^2 (accuracy) manually

formula for $R^2$:

$ R^2 = 1 - \frac{\Sigma(y - \hat{y})^2}{\Sigma(y-\bar{y})^2} $

where,
- $y$ = actual value
- $\hat{y}$ = predicted value
- $\bar{y}$ = mean of the actual value

```python
total_1, total_2 = 0, 0

y_mean = np.mean(y)

# Numerator
for i in range(len(y)):
    #find difference
    diff = y - y_pred[i]
    #square of the difference
    diff_sq = diff * diff
    #summation of the differences
    total_1 = total_1 + diff_sq
    
# Denominator
for i in range(len(y)):
    #find difference
    diff = y - y_mean
    #square of the difference
    diff_sq = diff * diff
    #summation of the differences
    total_1 = total_1 + diff_sq
    
my_R2 = 1 - total_1/total_2
```

```python
print(my_R2)
```
