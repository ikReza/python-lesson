#### Task-5: Add another column: `Stop Invert(m)`. Do necessary calculation and use the columns that is needed for this calculation. Take help from previous notebook if needed.

Formula:

$stopInvert = startInvert - \frac{slope}{1000}*conduitLength$

The table will be like below:

| |Conduit Label | Start Invert(m) |	Slope(m/km) | Conduit Length(m)| Stop Invert(m)|
|-|:-:|:-:|:-:|:-:|:-:|
|0|	CJAD4-253|	6.50|	2.67|	33.7|	6.410021|
|1|	CJAD4-186|	6.32|	1.50|	30.0|	6.275000|
|2|	CJAD4-15|	6.28|	1.95|	22.6|	6.235930|
|3|	CJAD4-183|	5.98|	1.25|	30.0|	5.942500|
|4|	CJAD4-163|	5.95|	1.25|	30.0|	5.912500|

I've Showed only first five rows above

````python
sk["Stop Invert(m)"]=sk["Start Invert(m)"]-((sk["Slope(m/km)"]/1000)*sk["Conduit Length(m)"])
````

````python
print(sk)
````
